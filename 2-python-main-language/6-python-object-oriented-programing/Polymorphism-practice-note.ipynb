## 3. Run-Time Polymorphism Through Method Overriding

```python
class Person:
    def greet(self):
        print("Hello")

class BritishPerson(Person):
    def greet(self):
        print("Good day, mate")

class FrenchPerson(Person):
    def greet(self):
        print("Bonjour")

person = Person()
brit = BritishPerson()
french = FrenchPerson()

person.greet()
brit.greet()
french.greet()
```

```python
class Product:
    """
    Parent class - Defines WHAT all products should be able to do
    Provides DEFAULT behavior that children can override
    """

    def __init__(self, name, price):
        self.name = name
        self.price = price

    def calculate_discount(self):
        """
        DEFAULT discount: No discount (0%)
        Children can override this to provide their own discounts
        """
        print(f"Using Product.calculate_discount() [PARENT VERSION]")
        final_price = self.price
        print(f"No discount: Â£{final_price:.2f}")
        return final_price

    def show_info(self):
        """
        Display basic product info
        This is NOT overridden - all children use this version
        """
        print(f"\nðŸ“¦ {self.name} - Â£{self.price:.2f}")
```

```python
class Book(Product):
    """
    Book products - OVERRIDES discount to give books 10% off
    """

    def calculate_discount(self):
        """
        OVERRIDDEN METHOD - Book-specific discount
        Books always get 10% off
        """
        print(f"Using Book.calculate_discount() [CHILD VERSION]")
        discount = self.price * 0.10
        final_price = self.price - discount
        print(f"10% Book Discount: Save Â£{discount:.2f}")
        print(f"Final Price: Â£{final_price:.2f}")
        return final_price


class Electronics(Product):
    """
    Electronics - OVERRIDES discount with price-based tiers
    """

    def calculate_discount(self):
        """
        OVERRIDDEN METHOD - Electronics-specific discount
        Expensive items get bigger discounts
        """
        print(f"Using Electronics.calculate_discount() [CHILD VERSION]")

        # Decide discount based on price
        if self.price > 500:
            discount_percent = 0.15  # 15% for expensive items
            tier = "Premium"
        elif self.price > 200:
            discount_percent = 0.10  # 10% for mid-range
            tier = "Mid-Range"
        else:
            discount_percent = 0.05  # 5% for budget items
            tier = "Budget"

        discount = self.price * discount_percent
        final_price = self.price - discount

        print(f"{int(discount_percent * 100)}% {tier} Discount: Save Â£{discount:.2f}")
        print(f"Final Price: Â£{final_price:.2f}")
        return final_price


class Clothing(Product):
    """
    Clothing - OVERRIDES discount with seasonal rules
    """

    def __init__(self, name, price, season):
        super().__init__(name, price)
        self.season = season  # 'summer' or 'winter'

    def calculate_discount(self):
        """
        OVERRIDDEN METHOD - Clothing-specific discount
        End-of-season sales: 30% off!
        """
        print(f"Using Clothing.calculate_discount() [CHILD VERSION]")

        import datetime
        current_month = datetime.datetime.now().month

        # Big discounts in opposite season
        if self.season == 'summer' and current_month in [11, 12, 1, 2]:
            discount_percent = 0.30
            reason = "End of Summer Sale (Winter Now!)"
        elif self.season == 'winter' and current_month in [5, 6, 7, 8]:
            discount_percent = 0.30
            reason = "End of Winter Sale (Summer Now!)"
        else:
            discount_percent = 0.05
            reason = "Standard Discount"

        discount = self.price * discount_percent
        final_price = self.price - discount

        print(f"{int(discount_percent * 100)}% {reason}")
        print(f"Save Â£{discount:.2f}")
        print(f"Final Price: Â£{final_price:.2f}")
        return final_price


class Grocery(Product):
    """
    Grocery - OVERRIDES discount based on expiry date
    """

    def __init__(self, name, price, expiry_days):
        super().__init__(name, price)
        self.expiry_days = expiry_days

    def calculate_discount(self):
        """
        OVERRIDDEN METHOD - Grocery-specific discount
        Items expiring soon get big discounts
        """
        print(f"Using Grocery.calculate_discount() [CHILD VERSION]")

        # Bigger discount for items expiring soon
        if self.expiry_days <= 2:
            discount_percent = 0.50
            urgency = "50% OFF - Expires in 1-2 Days!"
        elif self.expiry_days <= 5:
            discount_percent = 0.25
            urgency = "25% OFF - Expires in 3-5 Days"
        else:
            discount_percent = 0.10
            urgency = "10% OFF - Fresh!"

        discount = self.price * discount_percent
        final_price = self.price - discount

        print(f"{urgency}")
        print(f"Save Â£{discount:.2f}")
        print(f"Final Price: Â£{final_price:.2f}")
        return final_price
```

```python
def apply_discount(product):
    """
    THIS IS RUN-TIME POLYMORPHISM!

    When you write this function, you don't know what type
    'product' will be. Python figures it out WHEN the program runs.
    """
    print("\n" + "="*60)
    print("â° RUNTIME DECISION HAPPENING NOW...")
    print("="*60)

    # Python checks: "What type is this product?"
    print(f"ðŸ” Checking product type...")
    print(f"   Type found: {type(product).__name__}")
    print(f"   Will call: {type(product).__name__}.calculate_discount()")
    print("-"*60)

    product.show_info()

    # RUNTIME POLYMORPHISM: Python calls the correct method
    # based on the object's ACTUAL type
    final_price = product.calculate_discount()

    print("="*60)
    return final_price
```

```python
def process_shopping_cart(cart):
    """
    Process multiple products - each could be a different type!
    Python makes a RUNTIME DECISION for EACH item in the loop.
    """
    print("\n" + "ðŸ›’"*30)
    print("SHOPPING CART CHECKOUT")
    print("ðŸ›’"*30)
    print(f"\nYour cart has {len(cart)} items")
    print("Python will make a runtime decision for each item...\n")

    total = 0

    for i, product in enumerate(cart, 1):
        print(f"\n{'â”€'*60}")
        print(f"ITEM #{i}")
        print(f"{'â”€'*60}")
        print(f"â° Runtime Check: What type is this?")
        print(f"   Answer: {type(product).__name__}")
        print(f"   Action: Call {type(product).__name__}.calculate_discount()")

        product.show_info()
        item_price = product.calculate_discount()
        total += item_price

    print(f"\n{'â•'*60}")
    print(f"ðŸ§¾ CART SUMMARY")
    print(f"{'â•'*60}")
    print(f"Total Items: {len(cart)}")
    print(f"Grand Total: Â£{total:.2f}")
    print(f"{'â•'*60}\n")
```
