```python
from fin import (remove_null_values, 
                 format_currency, 
                 remove_duplicates, 
                 is_valid_amount)
```

```python
# Example: Clean transaction data
transactions = [100.50, None, 200.75, '', 150.00, 100.50]
```

```python
clean_data = remove_null_values(transactions)
print(f"Clean data: {clean_data}")
```

```python
# Remove duplicates
unique_transactions = remove_duplicates(clean_data)
print(f"Unique transactions: {unique_transactions}")
```

```python
# Format currency
for amount in unique_transactions:
    if is_valid_amount(amount):
        print(format_currency(amount))
```
